#### 1. Create Python Array or List

```python
# Python Array (List) Operations Step by Step

# 1. Create a Python list
python_list = [1, 2, 3, 4, 5]
print("Original List:", python_list)

# 2. Access elements
print("First element:", python_list[0])
print("Last element:", python_list[-1])

# 3. Slicing
print("Slice [1:4]:", python_list[1:4])

# 4. Append an element
python_list.append(6)
print("After append:", python_list)

# 5. Insert an element
python_list.insert(2, 99)
print("After insert at index 2:", python_list)

# 6. Remove an element by value
python_list.remove(99)
print("After remove 99:", python_list)

# 7. Pop an element by index
popped = python_list.pop(3)
print("Popped element at index 3:", popped)
print("After pop:", python_list)

# 8. Update an element
python_list[0] = 100
print("After update first element:", python_list)

# 9. List comprehension (add 2 to each item)
added_list = [x + 2 for x in python_list]
print("Add 2 to each item:", added_list)

# 10. Multiply each item by 3
multiplied_list = [x * 3 for x in python_list]
print("Multiply each item by 3:", multiplied_list)

# 11. Filter items greater than 5
filtered_list = [x for x in python_list if x > 5]
print("Items greater than 5:", filtered_list)

# 12. Sort the list
python_list.sort()
print("Sorted list:", python_list)

# 13. Reverse the list
python_list.reverse()
print("Reversed list:", python_list)

# 14. Length of the list
print("Length of list:", len(python_list))

# 15. Extend the list with another list
python_list.extend([7, 8])
print("After extend:", python_list)

# 16. Count occurrences of an element
count_2 = python_list.count(2)
print("Count of 2:", count_2)

# 17. Find index of an element
if 4 in python_list:
    index_4 = python_list.index(4)
    print("Index of 4:", index_4)

# 18. Clear the list
python_list.clear()
print("After clear:", python_list)

# Additional Python List Operations

# 19. Copy a list
python_list = [1, 2, 3, 4, 5]
copied_list = python_list.copy()
print("Copied list:", copied_list)

# 20. Nested lists (2D list access)
nested_list = [[1, 2], [3, 4]]
print("Element at [1][0]:", nested_list[1][0])

# 21. Remove all elements using slicing
python_list[:] = []
print("List after removing all elements with slicing:", python_list)

# 22. Using del to remove by index or slice
python_list = [1, 2, 3, 4, 5]
del python_list[2]
print("After deleting index 2:", python_list)
del python_list[:2]
print("After deleting first two elements:", python_list)

# 23. Repetition (multiply list)
repeated_list = [1, 2] * 3
print("Repeated list:", repeated_list)

# 24. Concatenate lists
concat_list = [1, 2] + [3, 4]
print("Concatenated list:", concat_list)
```
